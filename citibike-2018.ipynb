# MERGE CSV

```python
df1 = pd.read_csv("Resources/201806-citibike-tripdata.csv")
df2 = pd.read_csv("Resources/201807-citibike-tripdata.csv")
df3 = pd.read_csv("Resources/201808-citibike-tripdata.csv")
```

```python
data = [df1, df2, df3]
df = pd.concat(data).drop_duplicates()
```

```python
df.to_csv("Resources/2018-citibike-summer.csv", index=False)
```

```python
# Read in new csv
df = pd.read_csv("Resources/2018-citibike-summer.csv")
df.head()
```

```python
# Convert trip duration column from seconds to minutes
df["tripduration"] = [trip/60 for trip in df["tripduration"]]

# Rename trip duration column
df.rename(columns={"tripduration":"tripduration (min)"})

df.head()
```

```python
# Drop rows without a min of 2min and where the start and end stations are the same
# Filter out records where users rented a bike but found that it wasn't working properly
df = df.loc[((df["tripduration"]>2) & (df["start station id"] != df["end station id"])),:]
df.head()
```

```python
# Filter out trips longer than 24 hours 
df = df.loc[df["tripduration"] < 1440, :]
df.head()
```

```python
# Rename values in gender column with strings
gender = {0:"Unknown", 1:"Male", 2:"Female"}
df.gender = [gender[item] for item in df.gender]
df.head()
```
